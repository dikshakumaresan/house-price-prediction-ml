```python
columns_to_remove = ['yr_renovated', 'date', 'street', 'city', 'statezip', 'country']
df = df.drop(columns=columns_to_remove)
```

```python
import scipy.stats as stats

z_scores = stats.zscore(df)

threshold = 3

print("Size before removing outliers:", df.shape)

outliers_df = df[(z_scores > threshold).any(axis=1)]

df = df[(z_scores <= threshold).all(axis=1)]

print("Size after removing outliers:", df.shape)
```

```python
from sklearn.model_selection import train_test_split

from sklearn.neighbors import KNeighborsRegressor

from sklearn.linear_model import LinearRegression

from sklearn.tree import DecisionTreeRegressor

from sklearn.ensemble import RandomForestRegressor
```

```python
x= df.drop('price', axis=1)
y=df['price']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
models = [
    ('Random Forest', RandomForestRegressor()),
    ('Linear Regression', LinearRegression()),
    ('Decision Tree', DecisionTreeRegressor()),
    ('KNN', KNeighborsRegressor())
]
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
for name, model in models:
    print(name)
    print()

    model.fit(x_train, y_train)

    y_pred = model.predict(x_test)

    print("Mean Squared Error (MSE):", mean_squared_error(y_test, y_pred))
    print()

    print("Mean Absolute Error (MAE):", mean_absolute_error(y_test, y_pred))
    print()

    print("R-squared (R²):", r2_score(y_test, y_pred))
    print("\n")
```

```python
plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted House Prices")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate residuals
residuals = y_test - y_pred

# Create side-by-side plots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Plot 1: Actual vs Predicted
ax1.scatter(y_test, y_pred, alpha=0.5)

# Perfect prediction line
ax1.plot(
    [y_test.min(), y_test.max()],
    [y_test.min(), y_test.max()],
    'r--',
    linewidth=2
)

ax1.set_xlabel("Actual Prices")
ax1.set_ylabel("Predicted Prices")
ax1.set_title("Actual vs Predicted House Values")

# Plot 2: Residual Distribution
sns.histplot(residuals, kde=True, ax=ax2)

ax2.axvline(x=0, color='black', linestyle='--', linewidth=2)
ax2.set_xlabel("Residual (Actual - Predicted)")
ax2.set_title("Distribution of Residuals")

plt.tight_layout()
plt.show()
```
